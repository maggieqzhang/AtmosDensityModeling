```python
def time_series_data(hours, x_data, y_data):
    x = np.reshape(x_data[:hours,:],(1,hours,x_data.shape[1]))
    y = np.reshape(y_data[hours:hours+1,:], (1, 10))
    end = len(x_data)-hours-1
    for i in range(1,end):
        x = np.append(x, np.reshape(x_data[i:i+hours,:],(1,hours,x_data.shape[1])), axis =0)
        y = np.append(y,np.reshape(y_data[i+hours:hours+1+i,:], (1, 10)), axis = 0)
    return x,y
```
